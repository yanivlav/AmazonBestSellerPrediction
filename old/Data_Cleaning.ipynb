```python
import pandas as pd
import numpy as np
import json
import csv
```

```python
df = pd.read_csv("dataFrame_combined1.csv")
print(df.shape)

# Remove Corrupted Rows ---------------------------------------------------------------------
df = df.dropna(axis=0, thresh=8)
print(df.shape)


# Remove $ and , sign from Price ------------------------------------------------------------------
df["Price"] = df.Price.astype(str)
iprice = list()
iiprice = list()

for i in df.Price:
    iprice.append(i.replace("$", ""))
df.Price = iprice
for i in df.Price:
    iiprice.append(i.replace(",", ""))
df.Price = iprice
df["Price"] = df.Price.astype(float)
df = df.reset_index(drop=True)
df
```

```python
# Cleaning the Brand name ------------------------------------------------------------------

ibrand = list()

for i in df.Brand:
    ibrand.append(i.replace("Brand: ", ""))
df.Brand = ibrand
ibrand = list()
for i in df.Brand:
    ibrand.append(i.replace("Visit the ", ""))
df.Brand = ibrand
ibrand = list()
for i in df.Brand:
    ibrand.append(i.replace(" Store", ""))
df.Brand = ibrand
```

```python
# Cleaning the Ratings ------------------------------------------------------------------

iRating = list()
for i in df.Rating:
    iRating.append(i.replace(" ratings", ""))
df.Rating = iRating
iRating = list()
for i in df.Rating:
    iRating.append(i.replace(" rating", ""))
df.Rating = iRating
iRating = list()
for i in df.Rating:
    iRating.append(i.replace(",", ""))
df.Rating = iRating
```

```python
# Cleaning the Stars ------------------------------------------------------------------

iStars = list()
for i in df.Stars:
    iStars.append(i.split(' ')[0])
df.Stars = iStars
```

```python
# Cleaning the comments ------------------------------------------------------------------

iComments = list()
for i in df.Comments:
    if '+' in i:
        iComments.append(i.split('+')[0])
    else:
        iComments.append(i.split(' ')[0])
df.Comments = iComments

df = df.drop(df.columns[[0, 1]],axis = 1)
df
```
